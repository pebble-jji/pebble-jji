```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font = font_manager.FontProperties(fname = 'C:/Windows/Fonts/NanumGothic.ttf').get_name()
rc('font', family=font)

df = pd.read_csv('1911-2020 일별 기온.csv', encoding = 'cp949', index_col = 0)
```

```python
df.loc[df['최고기온(°C)'].isnull() | df['최저기온(°C)'].isnull() | df['평균기온(°C)'].isnull(), '지점명'].value_counts()
```

```python
df.일시 = pd.to_datetime(df.일시)
df.dropna(axis = 0, how = 'any', inplace = True)
df = df[df['일시'].dt.year == 2020]
```

```python
plt.figure(figsize = (15,3))
p = df[df.일시.dt.year == 2020].pivot_table(index = '지점명', values = '평균기온(°C)', aggfunc = 'mean').sort_values(by = '평균기온(°C)', ascending = False)
plt.bar(x = p.index, height = p['평균기온(°C)'], color = 'green')
plt.xticks(rotation=90, fontsize = 10)
plt.axhline(p['평균기온(°C)'].max(), color = 'red', ls = ':', alpha = 0.7, label = 'max')
plt.axhline(p['평균기온(°C)'].mean(), color = 'black', ls = ':', alpha = 0.7, label = 'mean')
plt.axhline(p['평균기온(°C)'].min(), color = 'blue', ls = ':', alpha = 0.7, label = 'min')
plt.title('지점별 일평균기온')
plt.legend()

plt.bar(x = '서울',height = p.loc[p.index =='서울', '평균기온(°C)'], color = 'blue')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (15,3))
p = df[df.일시.dt.year == 2020].pivot_table(index = '지점명', values = '최고기온(°C)', aggfunc = 'mean').sort_values(by = '최고기온(°C)', ascending = False)
plt.bar(x = p.index, height = p['최고기온(°C)'], color = 'red')
plt.xticks(rotation=90, fontsize = 10)
plt.axhline(p['최고기온(°C)'].max(), color = 'red', ls = ':', alpha = 0.7, label = 'max')
plt.axhline(p['최고기온(°C)'].mean(), color = 'black', ls = ':', alpha = 0.7, label = 'mean')
plt.axhline(p['최고기온(°C)'].min(), color = 'blue', ls = ':', alpha = 0.7, label = 'min')
plt.title('지점별 일최고기온')
plt.legend()

plt.bar(x = '서울',height = p.loc[p.index =='서울', '최고기온(°C)'], color = 'blue')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (15,3))
p = df[df.일시.dt.year == 2020].pivot_table(index = '지점명', values = '최저기온(°C)', aggfunc = 'mean').sort_values(by = '최저기온(°C)', ascending = False)
plt.bar(x = p.index, height = p['최저기온(°C)'], color = 'blue')
plt.xticks(rotation=90, fontsize = 10)
plt.axhline(p['최저기온(°C)'].max(), color = 'red', ls = ':', alpha = 0.7, label = 'max')
plt.axhline(p['최저기온(°C)'].mean(), color = 'black', ls = ':', alpha = 0.7, label = 'mean')
plt.axhline(p['최저기온(°C)'].min(), color = 'blue', ls = ':', alpha = 0.7, label = 'min')
plt.title('지점별 일최고기온')
plt.legend()

plt.bar(x = '서울',height = p.loc[p.index =='서울', '최저기온(°C)'], color = 'purple')

plt.tight_layout()
plt.show()
```
