# 32강. 인덱스, 컬럼명으로 데이터 추출하기

```python
import pandas as pd
df = pd.read_csv('C:/code/data/scores.csv', encoding = 'utf-8')
```

```python
# 데이터프레임.loc[row, col] : 인덱스와 컬럼명으로 데이터 추출
df.loc[ : 5, 'name']
```

```python
# 컬럼명으로 슬라이싱 가능
df.loc[ : 5, 'name' : 'kor']
```

```python
# row와 col 자리에 시퀀스 넣을 수 있음.
df.loc[[0,2,3,4,5], ['kor','eng']]
```

```python
# 조건을 넣어서 boolean indexing 가능
df.loc[df.kor > 95, :]
```

# 40강. 카테고리형 자료

괜히 문자나 숫자를 쓰는 것보다 용량과 속도면에서 좋을 수 있음.

```python
import numpy as np
df = pd.read_csv('C:/code/data/scores.csv', encoding = 'utf-8')
df['mean'] = df[['kor','eng','math']].apply(np.mean, axis = 1)
```

```python
def grade(x) : 
    if x >= 90 : 
        return 'A'
    elif x >= 80 : 
        return 'B'
    elif x >= 70 : 
        return 'C'
    elif x >= 60 : 
        return 'D'
    else : 
        return 'F'  
df['grade'] = df['mean'].apply(grade)
```

```python
df.info()
```

```python
df['grade'] = df['grade'].astype('category')
```

```python
df.info()
```
