# 64 ~ 65강.제목, 축 레이블, 그리드

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('C:/code/data/서울시 코로나19 확진자 현황.csv')
```

```python
# 결측치가 많은 컬럼 삭제
df.drop(['환자번호' , '국적', '환자정보', '조치사항','연번', '노출여부', '등록일','수정일','이동경로'], axis = 1, inplace = True)
```

```python
import matplotlib.font_manager as fm
[f.name for f in fm.fontManager.ttflist]
```

```python
df.확진일 = pd.to_datetime(df.확진일)
```

```python
# 꺾은선 그래프 : 항목간의 변화
df[df.지역 == '강동구'].pivot_table(
    index = '확진일', values = '지역', aggfunc = 'count'
    ).plot(marker = '.',    # marker : '*', '.' 등
           color = '#12C4E6', # color : hex 코드 가능
           linestyle = ':',
           ylim = (0, 50))
plt.grid(axis = 'both', color = 'black', alpha = 0.3)
plt.xticks(['2020-06', '2020-09', '2020-12', '2021-03', '2021-06','2021-09'], fontfamily = 'Gmarket Sans')
plt.title('강동구 코로나 확진자 변화 추이', loc = 'left', pad = 10, color = 'black', fontweight = 50, fontsize = 20, fontfamily = 'Gmarket Sans')
plt.show()
```

# 66강. 여러 개의 그래프와 범례

```python
gangdong = df[df.지역 == '강동구'].pivot_table(index = '확진일', values = '지역', aggfunc = 'count')
gwanak = df[df.지역 == '관악구'].pivot_table(index = '확진일', values = '지역', aggfunc = 'count')
plt.plot(gangdong['지역'], color = '#a1b3c3', marker = '.')
plt.plot(gwanak['지역'], color = '#000000', marker  = '.')

plt.xticks(rotation = 45)
plt.title('강동 vs 관악', loc = 'left', fontsize = 15, fontfamily = 'Gmarket Sans')
plt.xlabel('확진일', fontfamily = 'Gmarket Sans')
plt.ylabel('누적 확진자', fontfamily = 'Gmarket Sans')
plt.legend(['강동', '관악'])

plt.show()
```

# 71 ~ 77강. 여러가지 그래프

```python
# 데이터 / 판다스 불러오기

import pandas as pd

df = pd.read_csv('C:/code/data/공공자전거 대여이력 정보_2021.01.csv', encoding = 'cp949')

import matplotlib.pyplot as plt
 
# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font = font_manager.FontProperties(fname = 'C:/Windows/Fonts/NanumGothic.ttf').get_name()
rc('font', family=font)
```

```python
for col in list(df.columns) : 
    print(col)
    print(f'결측치 : {df[col].isnull().sum()} / {len(df)}')
    print(f"값 예시\n{list(df[col].unique())[0:5]}")
```

```python
df.info()
```

```python
# barplot

df['대여일시'] = pd.to_datetime(df['대여일시'])
df['반납일시'] = pd.to_datetime(df['반납일시'])

df['대여요일'] = df['대여일시'].dt.day_name().str[:3]
df['대여시간대'] = df['대여일시'].dt.hour
df['반납요일'] = df['반납일시'].dt.day_name().str[:3]
df['반납시간대'] = df['반납일시'].dt.hour

wd =  ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'] 

fig = plt.figure(figsize = (10,7))

for i in range(len(wd)) : 
    plt.subplot(2,4,i + 1)
    p = df[df['대여요일'] == wd[i]].pivot_table(index = '대여시간대', values = '자전거번호', aggfunc = 'count')
    plt.barh(p.index, p['자전거번호'], color = 'green')
    plt.title(f'시간대별 자전거 대여건수({wd[i]})', fontsize = 10)

fig.get_constrained_layout()
fig.suptitle('요일별 시간대별 자전거 대여건수', fontsize = 20,)

plt.show()

fig2 = plt.figure(figsize = (10,7))

for i in range(len(wd)) : 
    plt.subplot(2,4,i + 1)
    p = df[df['반납요일'] == wd[i]].pivot_table(index = '반납시간대', values = '자전거번호', aggfunc = 'count')
    plt.barh(p.index, p['자전거번호'], color = 'green')
    plt.title(f'시간대별 자전거 반납건수({wd[i]})', fontsize = 10)

fig2.get_constrained_layout()
fig2.suptitle('요일별 시간대별 자전거 반납건수', fontsize = 20,)

plt.show()
```

```python
# heatmap

import seaborn as sns

df['대여일시'] = pd.to_datetime(df['대여일시'])
df['반납일시'] = pd.to_datetime(df['반납일시'])

df['대여요일'] = df['대여일시'].dt.day_name().str[:3]
df['대여시간대'] = df['대여일시'].dt.hour
df['반납요일'] = df['반납일시'].dt.day_name().str[:3]
df['반납시간대'] = df['반납일시'].dt.hour

wd =  ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'] 

fig = plt.figure(figsize = (13,7))

for i in range(len(wd)) : 
    plt.subplot(2,4,i + 1)
    p = df[df['대여요일'] == wd[i]].pivot_table(index = '대여시간대', values = '자전거번호', aggfunc = 'count')
    sns.heatmap(p, cmap = 'Greens', cbar = False, xticklabels = ['대여 건수'])
    plt.title(f'시간대별 자전거 대여건수({wd[i]})', fontsize = 10)

fig.get_constrained_layout()
fig.suptitle('요일별 시간대별 자전거 대여건수', fontsize = 20)

plt.show()

fig2 = plt.figure(figsize = (13,7))

for i in range(len(wd)) : 
    plt.subplot(2,4,i + 1)
    p = df[df['대여요일'] == wd[i]].pivot_table(index = '반납시간대', values = '자전거번호', aggfunc = 'count')
    sns.heatmap(p, cmap = 'Greens', cbar = False, xticklabels = ['대여 건수'])
    plt.title(f'시간대별 자전거 반납건수({wd[i]})', fontsize = 10)

fig2.get_constrained_layout()
fig2.suptitle('요일별 시간대별 자전거 반납건수', fontsize = 20)

plt.show()
```
